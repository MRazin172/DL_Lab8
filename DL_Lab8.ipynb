```python
import os
import cv2
import numpy as np

# Path to the directory containing your original images
train = r'C:\Users\muzza\OneDrive\Desktop\Christ\DL\traffic\train\images'

# Path to the directory where augmented images will be saved
augmented_dir = r'C:\Users\muzza\OneDrive\Desktop\Christ\DL\traffic\train\augmented_images'

# Create the augmented directory if it doesn't exist
os.makedirs(augmented_dir, exist_ok=True)

# Function to apply data augmentation to an image
def augment_image(image):
    # Random rotation
    angle = np.random.randint(-15, 15)
    rows, cols = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (cols, rows))

    # Random scaling
    scale_factor = np.random.uniform(0.8, 1.2)
    scaled_image = cv2.resize(rotated_image, None, fx=scale_factor, fy=scale_factor)

    # Random flipping
    flip_direction = np.random.randint(-1, 2)  # -1 for horizontal flip, 0 for no flip, 1 for vertical flip
    flipped_image = cv2.flip(scaled_image, flip_direction)

    return flipped_image

# Iterate through each file in the original directory
for filename in os.listdir(train):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # Read the original image
        image_path = os.path.join(train, filename)
        original_image = cv2.imread(image_path)

        # Apply data augmentation
        augmented_image = augment_image(original_image)

        # Save the augmented image
        augmented_image_path = os.path.join(augmented_dir, filename)
        cv2.imwrite(augmented_image_path, augmented_image)
```
